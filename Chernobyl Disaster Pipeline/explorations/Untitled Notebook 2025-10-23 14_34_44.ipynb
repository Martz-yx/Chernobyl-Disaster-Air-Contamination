# The Chernobyl Disaster Air Contamination in neighbouring countries
The Chernobyl disaster was a nuclear accident that occurred on 26 April 1986 at the No. 4 nuclear reactor in the Chernobyl Nuclear Power Plant, near the city of Pripyat in the north of the Ukrainian SSR. It is considered the worst nuclear disaster in history and is one of only two nuclear energy disasters rated at seven - the maximum severity - on the International Nuclear Event Scale, the other being the 2011 Fukushima Daiichi nuclear disaster in Japan.

The immense release of **radioactive material** from the explosion and subsequent fires was carried by **air currents**, resulting in widespread contamination that extended far beyond the immediate vicinity of the plant. This report specifically focuses on the **transboundary impact** of the contamination. We will analyze the **effects of the accident on neighboring countries** by examining how the radioactive plume traveled and deposited contaminants across international borders. Due to the atmospheric spread, the contamination was not limited to neighboring nations but was also **detected in countries significantly far away from the Chernobyl site**. This fact underscores the global scale of the disaster's initial airborne impact.

This Report presents concentration of **Iodine-131 (I-131)**, **Caesium-134 (Cs-134)**  and **Caesium 137 (CS-137)**  as aerosol particles which were measured in specific location and date, providing concrete evidence of the disaster's reach and the resulting **environmental and public health concerns** in those affected nations. 

## Max Radionuclides Levels sampled per City
> The max value registered per city during the sampled time (1986-04-27 ~ 1986-08-04)

```python
import pandas as pd
import plotly.express as px

# Load data from the Spark table into a pandas DataFrame
df = spark.table("workspace.default.max_contamination_by_city").toPandas()

# Define radionuclides and their display properties
radionuclides = [
    ("max_Iodine_131", "Iodine-131", "Greens"),
    ("max_Caesium_134", "Caesium-134", "Oranges"),
    ("max_Caesium_137", "Caesium-137", "Reds")
]

for col, title, color_scale in radionuclides:
    # Remove NaN values for plotting
    plot_df = df[df[col].notna()]
    fig = px.scatter_mapbox(
        plot_df,
        lat="longitude",
        lon="lattitude",
        size=col,
        color=col,
        hover_name="City",
        color_continuous_scale=color_scale,
        size_max=30,
        zoom=1.5,
        center={"lat": plot_df["longitude"].mean(), "lon": plot_df["lattitude"].mean()},
        mapbox_style="carto-positron",
        title=f"Max {title} Contamination by City"
    )
    fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0}, showlegend=False, coloraxis_showscale=False)
    fig.show()
```

### Iodine 131
Short-term hazard (8 days) . It is highly volatile and easily inhaled. It concentrates in the thyroid gland, posing a significant risk of thyroid cancer.

### Caesium 134
Medium-term hazard (2 years). It behaves like potassium and thus is absorbed by the body and distributed throughout muscle tissue.

### Caesium 137
Long-term hazard (30 years), it is absorbed by muscle and other soft tissues. Responsible for residual contamination decades after the event.
